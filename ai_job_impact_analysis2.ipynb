```python
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: File path in Google Drive
file_path = '/content/drive/MyDrive/ai_job_impact.csv'

# Step 4: Load dataset
df = pd.read_csv(file_path)

# Step 5: Basic inspection
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nFirst 5 rows:")
display(df.head())

print("\nDataset info:")
df.info()

print("\nMissing values:")
print(df.isnull().sum())

# Step 6: Simple analysis
print("\nJob Status counts:")
print(df['Job_Status'].value_counts())

print("\nIndustry counts:")
print(df['Industry'].value_counts())

print("\nAverage Salary Before AI:", df['Salary_Before_AI'].mean())
print("Average Salary After AI:", df['Salary_After_AI'].mean())

print("\nAverage Productivity Change by Job Status:")
print(df.groupby('Job_Status')['Productivity_Change_%'].mean())

# Step 7: Visualizations
sns.set(style="whitegrid")

# Job Status Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Job_Status', hue='Job_Status', palette='Set2', legend=False)
plt.title('Job Status Distribution')
plt.show()

# AI Adoption Level Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='AI_Adoption_Level', hue='AI_Adoption_Level', palette='Set1', legend=False)
plt.title('AI Adoption Level Distribution')
plt.show()

# Salary Before vs Salary After
plt.figure(figsize=(8,5))
plt.scatter(df['Salary_Before_AI'], df['Salary_After_AI'], alpha=0.5)
plt.xlabel('Salary Before AI')
plt.ylabel('Salary After AI')
plt.title('Salary Before vs After AI')
plt.show()

# Productivity Change by Automation Risk
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Automation_Risk', y='Productivity_Change_%', hue='Automation_Risk', palette='coolwarm', legend=False)
plt.title('Productivity Change by Automation Risk')
plt.show()

# Correlation heatmap
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
